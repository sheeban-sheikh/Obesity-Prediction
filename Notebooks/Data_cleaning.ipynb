```python
def calculate_bmi(row):
    try:
        weight=float(row['Weight'])
        height=float(row['Height_cm'])/100
        return round(weight/(height**2),2)
    except:
        return None
df['BMI']=df.apply(calculate_bmi,axis=1)
df
```

```python
bmi=df.pop('BMI')
df.insert(3,'BMI',bmi)
df
```

```python
df['Weight'].unique()
```

```python
df['Weight']=pd.to_numeric(df['Weight'],errors='coerce')
df.fillna({'Weight':df['Weight'].mean()},inplace=True)
```

```python
df['Height_cm'].unique()
```

```python
df['Height_cm']=pd.to_numeric(df['Height_cm'],errors='coerce')
df.fillna({'Height_cm':df['Height_cm'].mean()},inplace=True)
```

```python
df['BMI'].unique()
```

```python
df['BMI']=pd.to_numeric(df['BMI'],errors='coerce')
df.fillna({'BMI':df['BMI'].mean()},inplace=True)
```

```python
df[df.duplicated()]
```

```python
mode_value=df['Obesity_Status'].mode()[0]
df.fillna({'Obesity_Status':mode_value},inplace=True)
df['Obesity_Status']=df['Obesity_Status'].str.strip().str.lower().replace('not_obese','not obese')
df['Obesity_Status']=df['Obesity_Status'].replace({'obese':1,'not obese':0}).astype(int)
```
